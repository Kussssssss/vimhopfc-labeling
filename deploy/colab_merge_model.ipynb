Notebook này lấy adapter `tranthaihoa/gemma_context`, gộp với base model, rồi đẩy bản
**fp16 hoàn chỉnh** lên Hugging Face để vLLM/Modal phục vụ được.

## Trước khi chạy
1. Mở notebook này trong **Google Colab** (File -> Upload notebook).
2. Bật GPU: **Runtime -> Change runtime type -> T4 GPU -> Save**.
3. Lấy **HF token quyền Write**: https://huggingface.co/settings/tokens
4. Dán token vào ô code thứ 2 (chỗ `hf_xxx`).

Sau khi chạy xong sẽ có repo mới: **`Kus669/gemma_context_merged`**.

```python
# Cai dat Unsloth (neu loi, xem https://github.com/unslothai/unsloth)
!pip install -q unsloth huggingface_hub
```

```python
HF_TOKEN = "hf_xxx"  # <-- DAN TOKEN WRITE CUA BAN VAO DAY

SOURCE_ADAPTER = "tranthaihoa/gemma_context"      # repo LoRA goc
TARGET_MERGED  = "Kus669/gemma_context_merged"    # repo merged tren tai khoan cua ban
```

```python
from unsloth import FastLanguageModel

# Load adapter (Unsloth tu keo base model ve)
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = SOURCE_ADAPTER,
    max_seq_length = 2048,
    load_in_4bit = True,
)

# Gop LoRA + base -> day ban fp16 hoan chinh len HF
model.push_to_hub_merged(
    TARGET_MERGED,
    tokenizer,
    save_method = "merged_16bit",
    token = HF_TOKEN,
)
print("DONE -> dung repo:", TARGET_MERGED)
```
